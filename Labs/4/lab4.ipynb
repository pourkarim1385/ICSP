```python
%%file p_8.c
#include <stdio.h>

long fib(long n){
    if (n <= 2)
        return 1;
    return fib(n - 1) + fib (n - 2);
}
int main(void){
    long int n;
    scanf("%ld", &n);
    printf("fib(n) = %ld\n", fib(n));
    return 0;
}
```

```python
%%cmd
gcc p_8.c -o p_8.exe
p_8.exe
4
```
