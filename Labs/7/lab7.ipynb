Differences Between **malloc() & calloc()**

| Feature                   | `malloc`                                                | `calloc`                                        |
| ------------------------- | ------------------------------------------------------- | ----------------------------------------------- |
| **Full form**             | Memory Allocation                                       | Contiguous Allocation                           |
| **Function signature**    | `void* malloc(size_t size)`                             | `void* calloc(size_t nitems, size_t size)`      |
| **Number of arguments**   | 1                                                       | 2                                               |
| **Memory initialization** | Does **not** initialize memory; contains garbage values | Initializes memory to **zero**                  |
| **Purpose**               | Allocates a single block of memory                      | Allocates multiple blocks and initializes       |
| **Return value**          | Pointer to allocated memory or `NULL` if failed         | Pointer to allocated memory or `NULL` if failed |
| **Performance**           | Slightly faster (no initialization)                     | Slightly slower (because of initialization)     |
| **Use case**              | When you don’t need zeroed memory                       | When you want memory to start with zeros        |


```python
%%file p_1.c
#include <stdio.h> 
#include <stdlib.h> 
int main() { 
    int* p = (int*)malloc(10 * sizeof(int)); 
    int i; 
    printf("P = 0x%p\n", p); 
    printf("****************\n"); 
    for (i = 0; i < 10; i++){ 
        printf("p%d = %d\n",i, *(p+i)); 
    } 
    free(p); 
    printf("****************\n"); 
    printf("P = 0x%p\n", p); 
    return 0; 
} 
```

```python
%%cmd
gcc p_1.c -o p_1.exe
p_1.exe
```

```python
%%file p_4.c
#include <stdio.h> 
#include <stdlib.h> 
#define ZERO 0 
void printer(int** array_of_int, int size1 , int size2) { 
    for (int i = 0; i < size1; i++){ 
        for (int j = 0; j < size2; j++) 
            printf("%d ", array_of_int[i][j]); 
        printf("\n"); 
    } 
} 
void initializer(int** array_of_int, int size1, int size2) { 
    int num= 1; 
    for (int i = ZERO; i < size1; i++) 
        for (int j = ZERO; j < size2; j++){ 
            array_of_int[i][j] = num; 
            num++; 
        } 
    } 
int main() { 
    int row, col, i; 
    int** A; 
    printf("Enter row and column:\n"); 
    scanf("%d %d", &row, &col); 
    A=(int**) malloc(row*sizeof(int)); /* 1. Complete this instruction */ 
    if (A == NULL)  
        exit(EXIT_FAILURE); 
    for(i = 0; i < row ; i++) { 
        A[i]=(int*) malloc(col*sizeof(int)); /* 2. Complete this instruction */ 
    if (A[i] == NULL)  
        exit(EXIT_FAILURE); 
    } /* Now you have a 2D integer array */ 
    initializer(A,row,col); /* initializing the array */ 
    printer(A,row,col); /* printing the array */ 

    /*Don't forget to free the allocated memory when you don't need it any more*/ 
    for (i = 0; i < row; i++) 
        free(A[i]);  
    free(A);  
return 0;
}
```
