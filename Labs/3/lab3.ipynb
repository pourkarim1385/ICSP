```python
%%file p_five.c
#include <stdio.h>
int main(void){
    int x =1, y=0, z=4;
    int a = --x && y || (z>>=2);
    printf("x= %d\ny= %d\nz= %d\na= %d\n", x,y,z,a);
    return 0;
}
```

```python
%%cmd
gcc p_five.c -o p_5.exe
p_5.exe
```

Let`s execute code step by step:

--x : pre-decerment; decrease value of x by 1 and return new value
-xx (which is 1 - 1 = 0) && (and) y (which has value of 0): Logically 0 and 0 is 0
z>>=2: shift (0100) by 2 bits (= 0001)
Logically 0 || (or) 1 is 1

some notes:

<HOW DOES COMPILER DEAL WITH THESE LOGICAL OPERATORS?>

--x &&:

we know that value of --x is 0 and basically 0 and anything is 0

As a result: compiler doesn`t check y and skip it and directly goes to ||(or)

(Also if it was 1 || compiler would skip this 1 or anything is 1)

but here is 0 so compiler cehcks (z>>=2) (which is 1) and return value of 1

```python
%%file p_six.c
#include <stdio.h>

int main(void){
    int x =1, y =0, z=4;
    int a = --x && y || (z>>=2);
    !a ? printf("%d\n", z) : printf("%d\n", ++z);
    return 0;
}
```

```python
%%cmd
gcc p_six.c -o p_six.exe
p_six.exe
```

--x (value of 1-1 = 0) and anything = =0
0 || (or) z>>=2 (1) is 1
So a is 1

!a reverse the truth value of a (from 1 to 0)

<Structure of if else condition in one line>

Condition ? statemnt_if_true :  statemnt_if_false

!a is 0 so compiler executre the "statemnt_if_false" statements

and print ++z which is 1 + 1 = 2 as we se the output in the console
