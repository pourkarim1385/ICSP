این برنامه درست اجرا نمیشود زیرا تابع ما متغییر هایی که در اسکوپ خودش هستند رو مرتب میکنه درواقع برای عملکرد درست باید متغیر های درون تابع مین را درست کنه

برای این کار دو روش وجود داره، تابع خروجی بدهد. یا با استفاده از پوینتر به متغیر های درون مین دسترسی پیدا کنیم و آنها را تغییر بدهیم

```python
%%file p_5.c
#include<stdio.h>
void num_sorter(int* first_num, int* second_num){
    if (*first_num< *second_num){
        int temp = *first_num;
        *first_num = *second_num;
        *second_num = temp;
    }
}
int main(){
    int first_num=5, second_num=7;
    num_sorter(&first_num, &second_num);
    printf("first= %d\nsecond= %d", first_num, second_num);
    return 0;
}
```
